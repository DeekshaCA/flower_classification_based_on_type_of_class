```python

import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split   
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

path="C:\\Users\\HP\\OneDrive\\Desktop\\projectML\\5_flowerclassrecognition\\Train.csv"
data=pd.read_csv(path)
print(data)
print(data.info())


inputs = data.drop(['Area_Code','Region_Code','Species','Class'],'columns')
output = data.drop(['Area_Code','Locality_Code','Region_Code','Height','Diameter','Species'],'columns')
# print(inputs)
# print(output)

# count=output['Class'].value_counts()
# print(count)

x_train,x_test,y_train,y_test = train_test_split(inputs,output,train_size=0.8)  #train_split : determains the training size i.e 80%
print("x_train",x_train)
print("x_test",x_test)
print("y_train",y_train)
print("y_test",y_test)

# count1=y_train['Class'].value_counts()
# print(count1)
# count2=y_test['Class'].value_counts()
# print(count2)

from sklearn.preprocessing import StandardScaler   # to increase the accuracy only use "StandardScaler" 
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

model = LogisticRegression() 
model.fit(x_train,y_train)    # training

y_pred = model.predict(x_test)
print(y_pred) 


from sklearn.metrics import confusion_matrix  # plot actual value with predicted value for checking accuracy of the model
cm = confusion_matrix(y_test,y_pred)
print(cm)

res = model.predict([[12,0,8]])
print(res)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
